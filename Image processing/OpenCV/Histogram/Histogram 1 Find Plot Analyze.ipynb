Histogram is a graph which gives a overall idea bout the intesnity distribution of an image.
X-axis is the pixels values and Y-axis is the summatio of pixels intensites


## Histogram calculation in openCV
cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('illustration.png',0)
#image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image,cmap='gray')
```

```python
hist=cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(hist)
```

## Histogram calculation in numpy
Numpy has another function, np.bincount() which is much faster than (around 10X) np.histogram(). So for one-dimensional histograms, you can better try that. Don’t forget to set minlength = 256 in np.bincount. For example, hist = np.bincount(img.ravel(),minlength=256)

```python
hist,bins=np.histogram(image.ravel(),256,[0,256])
plt.plot(hist)
```

## Plotting histogram using matplotlib

```python
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```python
image=cv2.imread('breakout.png')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
