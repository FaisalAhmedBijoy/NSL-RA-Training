```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread('Lena.png')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

## 1. Compute the gradient in the x and y direction

```python

gx=cv2.Sobel(gray,cv2.CV_32F,1,0)
gy=cv2.Sobel(gray,cv2.CV_32F,0,1)
magnitude,angle=cv2.cartToPolar(gx,gy)
f,(ax1,ax2)=plt.subplots(1,2,figsize=(20,20) )
ax1.imshow(gx,cmap='gray')
ax2.imshow(gy,cmap='gray')
```

## 2. Define the cells and bins

```python

n_bins=9
bins=np.int32(n_bins *angle / (2*np.pi) )
len(bins)
```

## 3 Calculate HOG Descriptor


```python
# parameter for define a HOG feature vector
win_size=(64,64)
block_size=(16,16)
block_stride=(5,5)
cell_size=(8,8)
n_bins=9

# Create the HOG descriptor
hog=cv2.HOGDescriptor(win_size,block_size,block_stride,cell_size,n_bins)

```

```python
hog.compute(image)
```
