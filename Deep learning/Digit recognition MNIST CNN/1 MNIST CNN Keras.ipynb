### MNIST
Dataset contains 70000 image with 28*28 pixels with 10 different class.
Keras split the dataset into two part
Train set : 60,000
Test set: 10,000

```python
import tensorflow as tf
```

```python
from tensorflow.keras.datasets import mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

## Analysis of MNIST dataset

```python
print('x_train shape',x_train.shape)
print('y_train shape',y_train.shape)
print('x_test shape',x_test.shape)
print('y_test shape',y_train.shape)
```

## Data normalization
Convert int to float32


```python
x_train=x_train.astype('float32')
x_test=x_test.astype('float32')

x_train/=255
x_test/=255
print(x_train)
```

```python
x_train=x_train.reshape(x_train.shape[0],28,28,1)
x_test=x_test.reshape(x_test.shape[0],28,28,1)
```

```python
print('New shape of dataset')
print('x_train shape',x_train.shape)
print('y_train shape',y_train.shape)
print('x_test shape',x_test.shape)
print('y_test shape',y_test.shape)
```

```python
print(x_train[0])
print(y_train[0])
```

## Convolutional Neural Network (CNN) Model 
Sequential API is the easier to create the Keras model
Adds layers sequentially to build the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense,Flatten,Dropout
```

```python
# Create the model
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=(5,5),activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(filters=32,kernel_size=(5,5),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(rate=0.25))

model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(rate=0.25))

model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

## Compile a model
For the training of a model, we have to define the learning process
we need to specify an optimizer, a loss function and optionally metrics like accuracy

the <b>loss function</b> is a measure on how good our model is at achieving the given objective
An <b>optimizer</b> is used to minimize the loss function by updateing the weight using the gradient


```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

## Train or Fit the model
an <b>epoch</b> refers to one cycle through the full training dataset. Mostly, training a neural network takes more than a few epochs
<b>batch size</b> is a number that is used to divided the dataset among the total dataset

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train=to_categorical(y_train,num_classes=10)
y_test=to_categorical(y_test,num_classes=10)
```

```python
y_train[:10]
```

```python
history=model.fit(x_train,y_train,epochs=5)
```

## Evaluating model
test the model using x_test and y_test

loss value, accuracy value =model.evaluate()

```python
score=model.evaluate(x_test,y_test)
print(score)
```

```python
print("Accuracy of the model is : ",score[1]*100)
```

## Visualizing the training process
we can visualize our training and testing accuracy and loss for each epoch so we can analyze our performance of our model.
the <b>accuracy value</b> and <b>loss values</b> are saved into the <b>history variable</b>
we can used <b>matplotlib</b> to plot the data

```python
plt.plot(history.history['accuracy'],label='training accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('Accuracy graph')
plt.legend()
```

```python
plt.plot(history.history['loss'],label='training loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Loss graph')
plt.legend()
```

## Save and load model

```python
import os
model_name='mnist_keras.h5'
model.save(model_name)
print('model save to: ',os.getcwd)
```
