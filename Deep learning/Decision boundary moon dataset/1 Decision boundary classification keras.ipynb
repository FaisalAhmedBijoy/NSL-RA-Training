## Moon type datasets

```python
from sklearn.datasets import make_moons
x,y= make_moons(n_samples=100,noise=0.1,random_state=0)
```

```python
x.shape, y.shape
```

```python
print(x[0])
print(y[0])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
plt.plot(x[y==0,0],x[y==0,1],'ob',alpha=0.5) # 'or'   # red circles
plt.plot(x[y==1,0],x[y==1,1],'xr',alpha=0.5)
plt.legend(['0','1'])
```

## Split the data train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

## Decision boundary using tensorflow 

```python
import tensorflow as tf
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(1,input_dim=2,activation='sigmoid')
])
```

```python
model.compile(tf.keras.optimizers.Adam(lr=0.05), 'binary_crossentropy',metrics='accuracy')
```

```python
history=model.fit(x_train,y_train,epochs=200)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Loss')
```

```python
plt.plot(history.history['accuracy'])
plt.xlabel('epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Graph')
```

```python
results=model.evaluate(x_test,y_test)
```

```python
print('Accuracy: ',results[1] * 100)
```

## mlxtend plotting using plot_decision_regions

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=x,y=y,clf=model,legend=2)
```

## Deep neural network Keras

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(4,input_shape=(2,),activation='tanh')
         )
model.add(tf.keras.layers.Dense(2,activation='tanh'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
model.compile(tf.keras.optimizers.Adam(lr=0.05),'binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_test,y_test,epochs=100)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Loss Graph')
```

```python
plt.plot(history.history['accuracy'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('Accuracy Graph')
```

```python
results=model.evaluate(x_test,y_test)
```

```python
print('Accuracy: ',results[1]*100)
```

```python
plot_decision_regions(X=x,y=y,clf=model,legend=2)
plt.xlabel('x',size=5)
plt.ylabel('y',size=5)
plt.title('Plot Decision boundary')
plt.show()
```
